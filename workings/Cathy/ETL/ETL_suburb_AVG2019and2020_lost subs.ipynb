# Extract - Melbourne House Data 

```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Reading in dataframe from CSV
df = pd.read_csv("Resources/Melbourne_housing_FULL.csv")
df.head()
```

# EDA AND TRANSFORM - Melbourne House Data 

```python
full_len = len(df)
full_len
```

```python
# Splitting the date
df[['Day', 'Month', 'Year']] = df["Date"].str.split("/", expand=True,)
df.head()
```

```python
# List of columns
df.columns
```

```python
# Reducing to columns of interest
# 'Landsize'
df = df.loc[:,['Suburb', 'Rooms', 'Type', 'Price', 'Bathroom', 'Car', 'Month', 'Year']]
df.head()
```

```python
# Changing column headers to lower case
df.columns = df.columns.str.lower()
df.head()
```

```python
# Investigating columns with null values
df.isnull().sum()
```

```python
# Dropping these null values
house_df = df.dropna() 
house_df.head()
```

```python
# Checking the length and that removal of na has worked
final_len = len(house_df)
print(final_len)
print(house_df.isnull().sum())
```

```python
difference = full_len - final_len
percent = (difference/full_len)*100
(f"The dataframe has been reduced from {full_len} rows to {final_len} rows, by {round(percent,2)} percent")
```

```python
# Investigating different types of houses
house_df['type'].unique()
```

# Extract - VIC Crime Data

```python
crime_vic_df = pd.read_csv("Resources/Data_Tables_LGA_Criminal_Incidents_Year_Ending_September_2020/Table 03-Table 1.csv")
crime_vic_df.head()
```

# EDA AND TRANSFORM - VIC Crime Data

```python
# Investigating columns
crime_vic_df.columns
```

```python
# This data is from 2020 and 2019 only
crime_vic_df['Year'].unique()
```

```python
# Selecting columns of interest
crime_vic_df = crime_vic_df.loc[:,['Suburb/Town Name', 'Year', 'Incidents Recorded']]
crime_vic_df.head()
```

```python
# Renaming columns
crime_vic_df.columns = 'suburb', 'year', 'incidents_recorded'
crime_vic_df.head()
```

```python
# Seperating 2019 incidents from 2020 
crime_2019_df = crime_vic_df.loc[(crime_vic_df["year"] == 2019), :]
crime_2019_df.head()
```

```python
# Changing col names
crime_2019_df = crime_2019_df.rename(columns = {"incidents_recorded" : "2019_incidents"})
crime_2019_df.head()
```

```python
# Dropping year column
crime_2019_df = crime_2019_df[["suburb", "2019_incidents"]]
crime_2019_df.head()
```

```python
crime_2019_df.dtypes
```

```python
# Removing , from numbers 
crime_2019_df['2019_incidents'] = crime_2019_df['2019_incidents'].str.replace(',', '')
# Converting from object to string to int
crime_2019_df['2019_incidents'] = crime_2019_df['2019_incidents'].astype(str).astype(int)
```

```python
crime_2019_df.dtypes
```

```python
total_2019 = crime_2019_df.groupby("suburb")["2019_incidents"].sum()
total_2019_df = pd.DataFrame({
    "total_2019_inc" : total_2019
})
total_2019_df.head()
```

```python
# Seperating 2020 incidents from 2019 
crime_2020_df = crime_vic_df.loc[(crime_vic_df["year"] == 2020), :]
crime_2020_df.head()
```

```python
# Changing col names
crime_2020_df = crime_2020_df.rename(columns = {"incidents_recorded" : "2020_incidents"})
crime_2020_df.head()
```

```python
# Dropping year column
crime_2020_df = crime_2020_df[["suburb", "2020_incidents"]]
crime_2020_df.head()
```

```python
crime_2020_df.dtypes
```

```python
# Removing , from numbers 
crime_2020_df['2020_incidents'] = crime_2020_df['2020_incidents'].str.replace(',', '')
# Converting from object to string to int
crime_2020_df['2020_incidents'] = crime_2020_df['2020_incidents'].astype(str).astype(int)
```

```python
crime_2020_df.dtypes
```

```python
total_2020 = crime_2020_df.groupby("suburb")["2020_incidents"].sum()
total_2020_df = pd.DataFrame({
    "total_2020_inc" : total_2020
})
total_2020_df.head()
```

```python
print(len(total_2020_df))
print(len(total_2019_df))
```

```python
# Merging 2019 and 2020 incidents together
crime_df = pd.merge(total_2020_df, total_2019_df, how = 'inner', on='suburb')
crime_df
```

```python
# Checking for null values 
crime_df.isnull().sum()
```

# Transform - merging house and crime data

```python
# Merging two dataframes on suburb
combined_df = pd.merge(house_df, crime_df, how = 'inner', on='suburb')
combined_df
```

```python
# Double checking for null values
combined_df.isnull().sum()
```

```python
combined_len = len(combined_df)
print(f"Combined df length: {combined_len}")
print(f"We have lost {final_len - combined_len} rows of housing data when merging the 2 dataframes")
```

```python
# Number of suburbs in housing dataframe
house_subs = house_df['suburb'].nunique()
# Number of suburbs in combined dataframe
combined_subs = combined_df['suburb'].nunique()
print(f"Number of suburbs in combined df: {combined_subs}")
print(f'We have lost a total of {house_subs - combined_subs} suburbs from merging housing data with crime data.')
```
