```python
print(df_no_sub['Distance'].min())
print(df_no_sub['Distance'].max())
```

```python
bins = [-1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
group_labels = ["0-5", "6-10", "11-15", "16-20", "21-25", "26-30", "31-35", "36-40", "41-45", "46-50"]
```

```python
# place the data series into a new column inside of the DataFrame
df_no_sub["kms_group"]=pd.cut(df_no_sub["Distance"], bins, labels=group_labels)
df_no_sub
```

```python
# grouping by distance
grouped_df = df_no_sub.groupby(["kms_group"])
distance_df = grouped_df.mean()
rounded_df = distance_df.round(0)
rounded_df
```

```python
rounded_df["Price"].tolist()
```
